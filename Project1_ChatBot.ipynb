```python
class ChatBot:
    def __init__(self, name="DecodeLabs Bot"):
        self.name = name
        self.is_running = True
        
        self.responses = {
            "hello"           : "Hello! How can I help you today?",
            "hi"              : "Hi there! What can I do for you?",
            "hey"             : "Hey! Nice to see you!",
            "how are you"     : "I am doing amazing! Thanks for asking!",
            "what is your name": "My name is DecodeLabs Bot! Created at DecodeLabs.",
            "who are you"     : "I am DecodeLabs Bot, a Rule-Based AI Chatbot!",
            "who created you" : "I was created by the team at DecodeLabs!",
            "tell me a joke"  : "Why do Python programmers wear glasses? Because they can't C!",
            "tell me another joke": "Why do programmers prefer dark mode? Light attracts bugs!",
            "what is ai"      : "AI is Artificial Intelligence - teaching computers to think like humans!",
            "what is python"  : "Python is a popular programming language used in AI and Machine Learning!",
            "what is machine learning": "Machine Learning is teaching computers to learn from data!",
            "thank you"       : "You are welcome! Always happy to help!",
            "thanks"          : "My pleasure! Anything else I can help with?",
            "help"            : "Try: hello, how are you, tell me a joke, what is ai, bye",
            "bye"             : "Goodbye! Have an amazing day!",
            "exit"            : "See you soon! Goodbye!",
            "quit"            : "Thanks for chatting! Goodbye!",
        }
    
    def get_response(self, user_input):
        user_input = user_input.strip().lower()
        
        if user_input in self.responses:
            response = self.responses[user_input]
            is_exit = user_input in ["bye", "exit", "quit"]
            return response, is_exit
        
        for key in self.responses:
            if key in user_input:
                response = self.responses[key]
                is_exit = key in ["bye", "exit", "quit"]
                return response, is_exit
        
        return "I don't understand that. Type 'help' to see commands!", False
    
    def run(self):
        print("\n" + "="*60)
        print("DecodeLabs AI Chatbot - Project 1")
        print("="*60)
        print(f"Bot: Hello! I am {self.name}.")
        print("Bot: Type 'help' for commands or 'bye' to exit.")
        print("-"*60 + "\n")
        
        while self.is_running:
            try:
                user_input = input("You: ")
                
                if not user_input.strip():
                    continue
                
                response, is_exit = self.get_response(user_input)
                print(f"Bot: {response}\n")
                
                if is_exit:
                    print("-"*60)
                    self.is_running = False
            
            except KeyboardInterrupt:
                print("\nBot: Goodbye!")
                break

bot = ChatBot("DecodeLabs Bot")
bot.run()
```
